# Boolean Indexing

🤔 What are the countries with more than one billion inhabitants?

```python
countries_df.head(2)
```

```python
big_countries = []

for i, c in countries_df.iterrows():
    if c['Population'] > 100_000_000:
        big_countries.append(c)
        
pd.DataFrame(big_countries).head(5)
```

```python
# In Pandas, this is a one-liner with Boolean Indexing:
countries_df[countries_df['Population'] > 100_000_000]
```

```python
countries_df[['Population']]
```

🤔 Which region of the world is the most populated?

```python
regions = countries_df.groupby('Region')
regions[['Population', 'Area (sq. mi.)']].sum()
```

```python
regions[['Population', 'Area (sq. mi.)']].sum().sort_values('Population', ascending=False)
```

# Plotting

```python
gdp = 'GDP ($ per capita)'

top_ten_countries_df = countries_df[[gdp]] \
    .sort_values(gdp, ascending=False) \
    .head(10)
```

```python
top_ten_countries_df
```

```python
top_ten_countries_df.plot(kind='bar');
```
